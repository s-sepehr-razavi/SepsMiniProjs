```python
to_split_on = ['\n', '،','!', ')','(','»',':','؟',' ','«', '\u200c', '\xa0','.']
```

```python
split_chars_pattern = '[' + re.escape(''.join(to_split_on)) + ']'

# Input string
def split_reverse(input_string):

    # Split the string based on the pattern
    split_result = re.split(split_chars_pattern, input_string)
    # split_result.reverse()

    l = []
    for word in split_result:
        if len(word) >= 2:
            l.append(word)

    return l
```

```python
def generate_data(data, c):    
    n = len(data)    

    X = []   
    Y = []        
    for i in range(n):        
        if i < c:
            context = data[:i] + data[i+1: i+c+1]
        elif i > n - 1 - c:
            context = data[i-c:i] + data[i+1:]
        else:
            context = data[i-c:i] + data[i+1:i+c+1]
        
        for word in context:
            X.append(word)
            Y.append(data[i])
    

    return X, Y
```

```python
# Custom Dataset
class SaebDataset(Dataset):
    def __init__(self, X, Y):        
        self.X = X
        self.Y = Y        

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        sample = self.X[idx]
        label = self.Y[idx]
        
        
        return sample, label
```

```python
dataset = SaebDataset(X, Y)
# Define the split sizes
train_size = int(0.8 * len(dataset))
eval_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - eval_size

# Split the dataset
train_dataset, eval_dataset, test_dataset = random_split(dataset, [train_size, eval_size, test_size])

# Create DataLoaders for each split
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
eval_loader = DataLoader(eval_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
class Skipgram(nn.Module):
    def __init__(self,vocab_size, hidden_size) -> None:
        super(Skipgram, self, ).__init__()  
        self.w1 = nn.Linear(vocab_size, hidden_size)
        self.w2 = nn.Linear(hidden_size, vocab_size)

    def forward(self, x):
        x = self.w1(x)
        x = self.w2(x)                

        return x
```

```python
max_iters = 1000
eval_interval = 100
learning_rate = 0.001
eval_iters = 100
```

```python
model = Skipgram(CORPUS_SIZE, 100)
model = model.to(device)
# print the number of parameters in the model
print(sum(p.numel() for p in model.parameters())/1e6, 'M parameters')

# create a PyTorch optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate)
loss_f = nn.CrossEntropyLoss()
```

```python
def estimate_loss(model, eval_iters, loss_f):
    out = {}
    model.eval()
    for split in ['train', 'val']:
        batch = next(enumerate(train_loader)) if split == 'train' else next(enumerate(eval_loader))
        losses = torch.zeros(eval_iters)
        for k in range(eval_iters):            
            X, Y = batch[1]
            # print(Y)
            logits = model(X)
            # print("here")
            # print((logits))
            # print(logits[0, Y])
            loss = loss_f(logits, Y)
            losses[k] = loss.item()
        out[split] = losses.mean()
    model.train()
    return out
```

```python
loss_hist = []
for iter in range(max_iters):

    # every once in a while evaluate the loss on train and val sets
    if iter % eval_interval == 0 or iter == max_iters - 1:
        losses = estimate_loss(model=model, eval_iters=eval_iters, loss_f= loss_f)
        loss_hist.append(losses)    
        print(f"step {iter}: train loss {losses['train']:.4f}, val loss {losses['val']:.4f}")

    # sample a batch of data
    xb, yb = next(enumerate(train_loader))[1] 

    # evaluate the loss
    logits = model(xb)
    loss = loss_f(logits, yb)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()
```
